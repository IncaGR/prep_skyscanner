```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from scipy import stats

import datetime as dt

from itertools import product
```

```python
np.random.seed(42)
```

```python
start_exp = pd.to_datetime("2026-02-16") 
delta_6_weeks = dt.timedelta(42)
# 6*7

dates_exp = pd.date_range(start=start_exp-delta_6_weeks,end=start_exp+delta_6_weeks)
print(len(dates_exp))

devices = ['mobile', 'desktop']

# mobile_users = 200
# dekstop_users = 400

mobile_users = np.arange(100, 201)   # 100 mobile users
desktop_users = np.arange(201, 301)  # 100 desktop users

# Assign treatment at user level — 40% treatment, 60% control
mobile_treatment = np.random.choice(
    ['treatment', 'control'],
    size=len(mobile_users),
    p=[0.4, 0.6]
)
mobile_group = dict(zip(mobile_users, mobile_treatment))
desktop_group = {user: 'control' for user in desktop_users}

# Generate sessions per user per day
sessions = []

for date in dates_exp:
    for user in mobile_users:
        # 30% chance the user visits on any given day
        if np.random.random() > 0.7:
            n_sessions = np.random.choice([1,2,3,4,5], p=[0.1,0.3,0.3,0.2,0.1])
            for _ in range(n_sessions):
                sessions.append({
                    'date': date,
                    'user_id': user,
                    'device': 'mobile',
                    'group': mobile_group[user]
                })
    for user in desktop_users:
        if np.random.random() > 0.7:
            n_sessions = np.random.choice([1,2,3,4,5], p=[0.1,0.3,0.3,0.2,0.1])
            for _ in range(n_sessions):
                sessions.append({
                    'date': date,
                    'user_id': user,
                    'device': 'desktop',
                    'group': desktop_group[user]
                })

fake_df = pd.DataFrame(sessions)
fake_df['session_id'] = np.arange(1, len(fake_df)+1)

print(f"Total sessions: {len(fake_df)}")
print(fake_df.head(10))
```
